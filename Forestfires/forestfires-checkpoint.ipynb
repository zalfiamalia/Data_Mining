```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("forestfires.csv")
data.head()
```

```python
data.info()
```

```python
data['month'].unique()
```

```python
latest_date = 'mar'
data[data['month']==latest_date]
```

```python
unique_month = data[data.RH>10]['month'].unique()
unique_month.sort()
unique_month
```

```python
month_rh = []
for i in unique_month:
  month_rh.append(data[data.RH>10][data['month']==i][data['month']==latest_date].area.sum())
```

```python
print(set(zip(unique_month, month_rh)))
```

```python
unique_day = data['day'][data.RH>10].unique()
unique_day
```

```python
day_rh = []
for i in unique_day:
  day_rh.append(data[data.RH>10][data['day']==i][data['month']==latest_date].area.sum())
```

```python
print(set(zip(unique_day, day_rh)))
```

```python
nan_indices = []
for i in range(len(unique_day)):
  if type(unique_day[i])==float:
    nan_indices.append(i)
```

```python
unique_day = list(unique_day)
day_rh= list(day_rh)
```

```python
for i in nan_indices:
  unique_day.pop(i)
  day_rh.pop(i)

unique_day
```

```python
plt.barh(unique_day, day_rh)
plt.show()
```

```python
plt.figure(figsize=(50,40))
plt.barh(unique_day, day_rh)
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.pie(month_rh)
plt.legend(unique_month, loc='best')
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.pie(day_rh)
plt.legend(unique_day, loc='best')
plt.show()
```
